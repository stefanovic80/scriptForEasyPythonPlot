```python
path = r"/home/stefano/Documents/Titanic" + "/"
files = os.listdir(path)
files

train = pd.read_csv(path + 'train.csv' )
"train"#train[Survived]
train.head()

test = pd.read_csv(path + 'test.csv' )
"test"
test.head()

g_submission = pd.read_csv(path + 'gender_submission.csv' )
"gender submission"
g_submission.head()
```

```python
fig = plt.figure(figsize = (13, 8))
ax1 = fig.add_subplot(221)
ax1.set_title('Males AND Females Survived')
train["Survived"].value_counts().plot.pie( autopct = "%.2f%%", colors = [ 'red', 'cyan' ] )
ax2 = fig.add_subplot(222)
ax2.set_title('Percetage of males vs females')
train['Sex'].value_counts().plot.pie( autopct = "%.2f%%", colors = [ 'cyan', 'pink' ] )

ax3 = fig.add_subplot(223)
ax3.set_title('Survived males')
train['Survived'][ train['Sex'] == 'male' ].value_counts().plot.pie( autopct = "%.2f%%", colors = [ 'red', 'cyan' ] )


ax4 = fig.add_subplot(224)
ax4.set_title('Survived females')
train['Survived'][ train['Sex'] == 'female' ].value_counts().plot.pie( autopct = "%.2f%%", colors = [ 'cyan', 'red' ] )
plt.show()
```
